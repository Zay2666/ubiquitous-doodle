## numerical variables

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def analyze_numeric(df):
    for column in df.select_dtypes(include=['number']).columns:
        print(f"Descriptive statistics for {column}:")
        print(df[column].describe())
        plt.figure(figsize=(10, 4))
        sns.histplot(df[column].dropna(), kde=True)
        plt.title(f"Histogram for {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

def analyze_categorical(df):
    for column in df.select_dtypes(include=['object']).columns:
        print(f"Value counts for {column}:")
        print(df[column].value_counts())
        plt.figure(figsize=(10, 4))
        sns.countplot(y=column, data=df)
        plt.title(f"Count Plot for {column}")
        plt.xlabel("Count")
        plt.ylabel(column)
        plt.show()



```

```python
analyze_numeric(df)
analyze_categorical(df)
```

## Bivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt


numeric_cols = df.select_dtypes(include=['number']).columns


corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Feature Correlation Matrix for Numeric Columns")
plt.show()

```

### comments: there is strong positive correlation between happiness index and GDP, Life expectancy, social support. Strong negative correlation between happiness index and corruption

```python


def bivariate_analysis_numeric(df, target_column):
    
    numeric_columns = df.select_dtypes(include=['number']).columns
    numeric_columns = numeric_columns.drop(target_column)  # 排除目标变量自身

    for column in numeric_columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df, x=column, y=target_column)
        plt.title(f'Scatter Plot between {target_column} and {column}')
        plt.xlabel(column)
        plt.ylabel(target_column)
        plt.show()

def bivariate_analysis_categorical(df, target_column):
   
    categorical_columns = df.select_dtypes(include=['object']).columns

    for column in categorical_columns:
        plt.figure(figsize=(12, 6))
        sns.boxplot(data=df, x=column, y=target_column)
        plt.title(f'Boxplot of {target_column} across different {column}')
        plt.xlabel(column)
        plt.ylabel(target_column)
        plt.xticks(rotation=45)  # 如果类别名称太长，可以旋转以便更好地显示
        plt.show()

bivariate_analysis_numeric(df, 'happiness')
bivariate_analysis_categorical(df, 'happiness')



```

### Comments：
01 Ignore the unnamed and the year rows, meaningless here;

02 Corruption mainly distributed from 0.6 - 1.0. The happiness index is widely distributed. It shows that the level of happiness does not depend entirely on corruption, and other factors need to be taken into account. And in corruption less than 0.2, there is also a very low happiness value, which may be an outlier.

03 The highest happiness index is distributed mainly in GPD capital from 40,000 to 60,000

04 The highest happiness index is mainly distributed in life expectancy from 70 to 75

05 Countries with population more than 1.2, the happiness is under 6, cannot be high happiness

06 The high happiness index is particularly concentrated in the range of 0.9 to 1 with high social support. But there are three obvious points in social support closest to 1, and the country at the top need to be analyzed separately to see why the happiness index has not improved a little more（might be an outlier), while the bottom two countries with low happiness index may be outliers

07 The happiness index of countries with high military expense is higher and more concentrated, and there some outliers

08 The happiness index is more widely distributed in Western Asia and northern Europe





## Multivariate Analysis

```python

def pairplot_analysis(df, vars_to_include, target_column):
   
    sns.pairplot(df, vars=vars_to_include, hue=target_column, plot_kws={'alpha': 0.5})
    plt.show()


from mpl_toolkits.mplot3d import Axes3D

def three_d_scatter(df, x_var, y_var, z_var, target_column):
    
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    xs = df[x_var]
    ys = df[y_var]
    zs = df[z_var]
    c = df[target_column] 
    
    img = ax.scatter(xs, ys, zs, c=c, cmap=plt.hot())
    fig.colorbar(img)
    ax.set_xlabel(x_var)
    ax.set_ylabel(y_var)
    ax.set_zlabel(z_var)
    plt.title('3D Scatter plot')
    plt.show()

vars_to_include = ['gdp_capita', 'social_support', 'life_expectancy']
target_column = 'happiness'
pairplot_analysis(df, vars_to_include, target_column)


x_var = 'gdp_capita'
y_var = 'freedom'
z_var = 'positive_affect'
target_column = 'happiness'
three_d_scatter(df, x_var, y_var, z_var, target_column)
```

### comments ： The plot of all three variables against the happiness index shows a dense area of data points, which may indicate that most countries are relatively concentrated on these indicators, while some countries show extreme values.
